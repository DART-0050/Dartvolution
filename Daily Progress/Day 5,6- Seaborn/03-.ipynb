SCATTERPLOT



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("C:/Users/Admin/Documents/GitHub/Dartvolution/Daily Progress/Day 5,6- Seaborn/mart_linePlot.xlsx").sample(50)
mart["Tier"]=mart['Outlet_Location_Type'].str.split(" ").str[1]
mart.head()
```

```python
sns.scatterplot(data=mart,x="Item_MRP",y="Sales",hue="Outlet_Size")
```

```python
sns.scatterplot(data=mart,x="Item_MRP",y="Sales",style="Outlet_Size",palette="husl",markers={"High":"^","Small":"v","Medium":"o"})
```

```python
sns.scatterplot(data=mart,x="Item_MRP",y="Sales",style="Outlet_Size",hue="Outlet_Location_Type",markers={"High":"^","Small":"v","Medium":"o"})
```

```python
sns.scatterplot(data=mart,
x="Item_MRP",
y="Sales",
style="Outlet_Size",
hue="Outlet_Size",
size="Tier",
sizes=(10,100), # size of markers left is min and right is max
palette="Purples_r",
markers={"High":"^","Small":"v","Medium":"o"}   )
```

LINE PLOT



```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales")
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",errorbar=None,estimator="sum")
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",errorbar="sd",estimator="sum") #ci=confidence interval also called errorbar
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",errorbar=('ci',20),n_boot=500) 
# means error bar is 20% of the data
# error bar is the standard error of the mean
# bootstarp is the number of times the data is sampled
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",
errorbar=None,
style="Outlet_Size",
markers=True,
dashes=True,
size="Outlet_Size")
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",
errorbar=None,
style="Outlet_Size",
hue="Tier")
```

REL PLOT



```python
sns.relplot(data=mart,x="Item_MRP",y="Sales",
hue="Outlet_Size",
col="Outlet_Size",
row="Tier",
kind="line")
```
