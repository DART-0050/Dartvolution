# Regression Plots



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.regplot(data=tips,
            x="total_bill",
            y="tip",
            color="r",
            marker="*",
            line_kws={"color":"black","ls":"--"},
            scatter_kws={"s":100,"color":"purple","alpha":0.5},
            ci=68,
            n_boot=500)
```
