### HEATMAPS
(Vales are depicted by colour)



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mart = pd.read_excel("C:/Users/Admin/Documents/GitHub/Dartvolution/Daily Progress/Day 5,6- Seaborn/mart_linePlot.xlsx")
mart.columns=mart.columns.str.lower()
mart.head()
```

```python
martpiv = mart.pivot_table(index="outlet_year", columns="outlet_size", values="sales")
martpiv
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15, 8))
sns.heatmap(martpiv,square=True,annot=True,fmt='.0f',
annot_kws={'size':15,"weight":'bold'},
linecolor='white',linewidths=1,
cmap='viridis')
```

```python
sns.heatmap(martpiv.corr(),vmin=-1,vmax=1,center=0)
# correlation matrix, vmin/vmax set to -1/1 to ensure all colors are used, center=0 to ensure the colorbar is centered
```

### CLUSTER MAPS



```python
sns.clustermap(martpiv,
col_cluster=False,
row_cluster=True,
annot=True,
fmt=".2f",
standard_scale=1)  # Standard scale = 1 standardizes columns else if 0 standardizes rows
```

```python
sns.clustermap(martpiv,
col_cluster=False,
row_cluster=True,
annot=True,
fmt=".2f",
z_score=0)  # z_score=0 normalizes the data  
```
