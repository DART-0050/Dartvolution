## FACETGRID PLOT



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mart = pd.read_excel("C:/Users/Admin/Documents/GitHub/Dartvolution/Daily Progress/Day 5,6- Seaborn/mart_linePlot.xlsx")
mart.columns=mart.columns.str.lower()
mart.head()
```

```python
mart.outlet_size.unique()
```

```python
mart.outlet_location_type.unique()
```

```python
l = sns.FacetGrid(mart,col="outlet_size")
```

```python
l = sns.FacetGrid(mart,col="outlet_size")
l.map_dataframe(sns.lineplot,x="outlet_year",y="sales")
```

```python
l = sns.FacetGrid(mart,col="outlet_size")
l.map_dataframe(sns.scatterplot,x="outlet_year",y="sales",hue="outlet_location_type",
marker="s",s=100,alpha=0.5,palette="deep")
```

```python
plt.figure(figsize=(20,10))

l = sns.FacetGrid(mart,col="outlet_size",row="outlet_location_type",sharey=True)
l.map_dataframe(sns.histplot,"sales")
l.set_axis_labels("Sales","Count")
l.set_titles("{col_name} Size | {row_name} Type")
l.tight_layout()
```
