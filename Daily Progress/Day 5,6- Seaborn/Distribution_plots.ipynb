```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
mart = pd.read_excel("C:/Users/Admin/Documents/GitHub/Dartvolution/Daily Progress/Day 5,6- Seaborn/supermarket_sales.xlsx")
mart.head()
```

## DISTRIBUTION PLOTS


HISTOGRAM



```python
sns.histplot(data=mart,x="Total",binwidth=50,element="step",fill=False)
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data=mart,x="Total",bins=np.arange(0,1100,50),kde=True,shrink=0.7)  #start,stop,step
plt.xticks(np.arange(0,1100,50))
```

```python
sns.histplot(data=mart,x="Total",hue="Payment",palette="Blues")
```

```python
sns.histplot(data=mart,x="Total",hue="Payment",palette="Blues",multiple="stack")
```

```python
sns.histplot(data=mart,x="Total",hue="Payment",binwidth=50,element="poly")
```

```python
sns.histplot(data=mart,x="Total",stat="probability")
```

```python
sns.histplot(data=mart,x="Total",y="gross income")
```

KDE (KERNEL DENSITY ESTIMATOR PLOT)



```python
sns.kdeplot(data=mart, x="Total")
```

```python
sns.kdeplot(data=mart, x="Total",hue="Gender",bw_adjust=0.5) #bandwidth
```

```python
sns.kdeplot(data=mart, x="Total",hue="Gender",
            multiple="stack",
            linewidth=2,
            palette="Dark2",
            alpha=0.1,
            bw_adjust=0.5) #bandwidth
```

```python
sns.kdeplot(data=mart, x="Total",log_scale=True) 
```

```python
sns.kdeplot(data=mart, x="Unit price",y="gross income",hue="Gender",fill=True,palette="Greens",levels=5,thresh=0.05)
```
